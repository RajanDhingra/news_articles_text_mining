Dropping design due to under-representation & popular due to ambiguous/non-objectifiable category

```python
data_filtered = data.loc[~data['category'].isin(['design','popular'])]
```

```python
data_filtered['read_time'].replace(regex=True,inplace=True,to_replace=r' min read',value=r'')
```

```python
data_filtered.drop(data_filtered.columns[0],axis=1,inplace=True)
```

```python
data_filtered.head()
```

```python
data_filtered[data_filtered.read_time.isnull()].read_time = '1'
```

```python
data_filtered.read_time.fillna('1',inplace=True)
```

```python
data_filtered.read_time.isnull().sum()
```

```python
data_filtered.read_time = data_filtered.read_time.astype(float)
```

```python
plt.bar(height = list(data_filtered.category.value_counts()),x=list(data_filtered.category.value_counts().index))
plt.show()
```

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
```

```python
lb = LabelEncoder()
data_filtered.category = lb.fit_transform(data_filtered.category)
```

```python
data_filtered.category = data_filtered.category.astype('category')
```

```python
Y = data_filtered.category
X = data_filtered.drop('category',axis=1)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.3, random_state = 0)
```

```python
plt.bar(height = list(Y_train.value_counts()),x=list(Y_train.value_counts().index))
plt.show()
```

```python
pd.concat([X_train,Y_train],axis=1).to_csv("C:/Users/Sabrish/Desktop/Data/train.csv")
```

```python
pd.concat([X_test,Y_test],axis=1).to_csv("C:/Users/Sabrish/Desktop/Data/test.csv")
```
