```python
import pandas as pd
import numpy as np
import urllib
import urllib.request
from bs4 import BeautifulSoup
```

```python
def unique(it):
        s = set()
        for el in it:
            if el not in s:
                s.add(el)
                yield el
```

```python
import time

from selenium import webdriver
from selenium.webdriver.common.keys import Keys

browser = webdriver.Chrome("/home/rajan/workshop/chromedriver")

browser.get("https://medium.com/topic/self")
time.sleep(1)

elem = browser.find_element_by_tag_name("body")

no_of_pagedowns = 2000

while no_of_pagedowns:
    elem.send_keys(Keys.PAGE_DOWN)
    print(no_of_pagedowns)
    time.sleep(0.1)
    no_of_pagedowns-=1
    
print("printed all links")
```

```python
data = browser.page_source
soup = BeautifulSoup(data, "html.parser")
```

```python
links = []
texts = []
```

```python
text=[]
i=0
for link in soup.find_all("h3", {"class": "bu bv bw w bx x ck ea eb d ct ec ed"}): 
    links.append("https://medium.com" + link.find("a").attrs["href"].split('?',1)[0])
    texts.append(link.find("a").text)
links_titles = pd.DataFrame(list(zip(links,texts)),columns=['Links','Titles'])
```
