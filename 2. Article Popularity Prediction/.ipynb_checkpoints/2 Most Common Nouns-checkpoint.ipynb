```python
import spacy
import pandas as pd
file = 'articles.csv'
data = pd.read_csv(file)
```

```python
data.head()
```

```python
nlp = spacy.load('en')
```

```python
from collections import Counter
min_token_length = 3
def isNoise(token):     
    is_noise = False
    if token.is_punct == True:
        is_noise = True 
    elif token.is_stop == True:
        is_noise = True
    elif token.pos_ != 'NOUN':
        is_noise = True
    elif len(token.string) <= min_token_length:
        is_noise = True
    return is_noise 
def cleanup(token, lower = True):
    if lower:
       token = token.lower() 
    return token.strip()
```

```python
#Most common nouns of each article
import sys
l = []
for i in data.index:
    doc = nlp(data['content'][i])
    cleaned_list = [cleanup(word.lemma_) for word in doc if not isNoise(word)]
    key = Counter(cleaned_list) .most_common(1)[0][0]
    share = data['shares'][i]
    l.append([key,share])
    sys.stdout.write("Articles : %d / %d \r" % (i,len(data)) )  
    sys.stdout.flush()
```

```python
df = pd.DataFrame(data=l)
```

```python
#calculate avg shares for each key noun
d_rows=df[df[0].duplicated(keep=False)]
df.drop(d_rows.index,axis=0,inplace=True)
items=d_rows.groupby(0).mean()
items[0]=items.index
df=df.append(items)
df.sort_values(by=1,inplace=True,ascending=False)
```

```python
df.to_csv('most_common_shares.csv',encoding = 'utf_8_sig',index=False)
```
